# RGB TO GRAY

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('bahan/lenna.jpg')  # Ganti 'your_image.jpg' dengan nama gambar Anda

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# Reduce to 1-bit grayscale (black and white)
gray_1bit = np.where(gray > 127, 255, 0).astype(np.uint8)  # Threshold at 127
print(image.shape)
print(gray.shape)
print(gray_1bit.shape)


# Convert to grayscale

# Plot RGB to Grayscale
plt.figure(figsize=(8, 4))
plt.subplot(131)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original (RGB)')
plt.subplot(132)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale 8 bit')
plt.subplot(133)
plt.imshow(gray_1bit, cmap='gray')
plt.title('Grayscale 1 bit')
plt.show()

```

# RGB TO HSV

```python

# Convert to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
print(hsv.shape)
# Plot RGB to HSV
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original (RGB)')
plt.subplot(122)
plt.imshow(hsv)
plt.title('HSV')
plt.show()
```

# RGB TO LAB

```python
# Convert to LAB
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

# Plot RGB to LAB
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original (RGB)')
plt.subplot(122)
plt.imshow(lab)
plt.title('LAB')
plt.show()
```

# RGB TO YUV

```python
# Convert to YUV
yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

# Plot RGB to CMYK
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original (RGB)')
plt.subplot(122)
plt.imshow(yuv)
plt.title('YUV')
plt.show()
```

# Greyscale TO RGB

```python
# Load grayscale image
gray_image = cv2.imread('bahan/lenna.jpg', cv2.IMREAD_GRAYSCALE)  # Ganti 'your_grayscale_image.jpg' dengan nama gambar grayscale Anda
print(gray_image.shape)
# Convert grayscale to RGB
rgb_image = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2RGB)
print(rgb_image.shape)

# Plot grayscale and RGB images
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale')
plt.subplot(122)
plt.imshow(rgb_image)
plt.title('RGB')
plt.show()
```
