```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load gambar
image = cv2.imread('bahan/lenna.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Transformasi Fourier
# Tujuan: Mengubah gambar dari domain spasial (pixel) ke domain frekuensi (komponen frekuensi).

f = np.fft.fft2(gray)
fshift = np.fft.fftshift(f)


# Memisahkan atau mengubah komponen frekuensi gambar.
# Jenis:
# Low-pass filter: Menghilangkan frekuensi tinggi (noise, detail halus).
# High-pass filter: Menghilangkan frekuensi rendah (blur, tepi).
# Band-pass filter: Menghilangkan frekuensi di luar rentang tertentu
# Buat filter low-pass
rows, cols = gray.shape
crow, ccol = rows // 2, cols // 2
mask = np.zeros((rows, cols), np.uint8)
radius = 50
cv2.circle(mask, (crow, ccol), radius, 1, -1)

# Terapkan filter pada domain frekuensi
fshift = fshift * mask

# Inverse Transformasi Fourier balikin ke domain gambar
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

# Normalisasi intensitas
img_back = cv2.normalize(img_back, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)

# Tampilkan hasil dengan Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_back, cmap='gray')
plt.title('Filtered')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load gambar
image = cv2.imread('bahan/lenna.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Transformasi Fourier
f = np.fft.fft2(gray)
fshift = np.fft.fftshift(f)

# Ukuran gambar
rows, cols = gray.shape
crow, ccol = rows//2, cols//2

# Buat filter low-pass
mask_lowpass = np.zeros((rows, cols), np.uint8)
radius_lowpass = 50
cv2.circle(mask_lowpass, (crow, ccol), radius_lowpass, 1, -1)

# Buat filter high-pass
mask_highpass = np.ones((rows, cols), np.uint8)
radius_highpass = 50
cv2.circle(mask_highpass, (crow, ccol), radius_highpass, 0, -1)

# Buat filter band-pass
mask_bandpass = np.zeros((rows, cols), np.uint8)
radius_inner = 0
radius_outer = 15
cv2.circle(mask_bandpass, (crow, ccol), radius_inner, 1, -1)
cv2.circle(mask_bandpass, (crow, ccol), radius_outer, 0, -1)

# Terapkan filter pada domain frekuensi
fshift_lowpass = fshift * mask_lowpass
fshift_highpass = fshift * mask_highpass
fshift_bandpass = fshift * mask_bandpass

# Inverse Transformasi Fourier
f_ishift_lowpass = np.fft.ifftshift(fshift_lowpass)
img_back_lowpass = np.fft.ifft2(f_ishift_lowpass)
img_back_lowpass = np.abs(img_back_lowpass)

f_ishift_highpass = np.fft.ifftshift(fshift_highpass)
img_back_highpass = np.fft.ifft2(f_ishift_highpass)
img_back_highpass = np.abs(img_back_highpass)

f_ishift_bandpass = np.fft.ifftshift(fshift_bandpass)
img_back_bandpass = np.fft.ifft2(f_ishift_bandpass)
img_back_bandpass = np.abs(img_back_bandpass)

# Normalisasi intensitas
img_back_lowpass = cv2.normalize(img_back_lowpass, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)
img_back_highpass = cv2.normalize(img_back_highpass, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)
img_back_bandpass = cv2.normalize(img_back_bandpass, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)

# Tampilkan hasil dengan Matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img_back_lowpass, cmap='gray')
plt.title('Low-Pass')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(img_back_highpass, cmap='gray')
plt.title('High-Pass')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(img_back_bandpass, cmap='gray')
plt.title('Band-Pass')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load gambar
image = cv2.imread('bahan/lenna.jpg')

# Split ke channel RGB
b, g, r = cv2.split(image)

# Transformasi Fourier untuk setiap channel
f_b = np.fft.fft2(b)
f_g = np.fft.fft2(g)
f_r = np.fft.fft2(r)

fshift_b = np.fft.fftshift(f_b)
fshift_g = np.fft.fftshift(f_g)
fshift_r = np.fft.fftshift(f_r)

# Ukuran gambar
rows, cols = b.shape
crow, ccol = rows//2, cols//2

# Buat filter low-pass
mask_lowpass = np.zeros((rows, cols), np.uint8)
radius_lowpass = 50
cv2.circle(mask_lowpass, (crow, ccol), radius_lowpass, 1, -1)

# Buat filter high-pass
mask_highpass = np.ones((rows, cols), np.uint8)
radius_highpass = 50
cv2.circle(mask_highpass, (crow, ccol), radius_highpass, 0, -1)

# Buat filter band-pass
mask_bandpass = np.zeros((rows, cols), np.uint8)
radius_inner = 0
radius_outer = 15
cv2.circle(mask_bandpass, (crow, ccol), radius_inner, 1, -1)
cv2.circle(mask_bandpass, (crow, ccol), radius_outer, 0, -1)

# Terapkan filter pada domain frekuensi untuk setiap channel
fshift_lowpass_b = fshift_b * mask_lowpass
fshift_highpass_b = fshift_b * mask_highpass
fshift_bandpass_b = fshift_b * mask_bandpass

fshift_lowpass_g = fshift_g * mask_lowpass
fshift_highpass_g = fshift_g * mask_highpass
fshift_bandpass_g = fshift_g * mask_bandpass

fshift_lowpass_r = fshift_r * mask_lowpass
fshift_highpass_r = fshift_r * mask_highpass
fshift_bandpass_r = fshift_r * mask_bandpass

# Inverse Transformasi Fourier untuk setiap channel
f_ishift_lowpass_b = np.fft.ifftshift(fshift_lowpass_b)
img_back_lowpass_b = np.fft.ifft2(f_ishift_lowpass_b)
img_back_lowpass_b = np.abs(img_back_lowpass_b)

f_ishift_highpass_b = np.fft.ifftshift(fshift_highpass_b)
img_back_highpass_b = np.fft.ifft2(f_ishift_highpass_b)
img_back_highpass_b = np.abs(img_back_highpass_b)

f_ishift_bandpass_b = np.fft.ifftshift(fshift_bandpass_b)
img_back_bandpass_b = np.fft.ifft2(f_ishift_bandpass_b)
img_back_bandpass_b = np.abs(img_back_bandpass_b)

f_ishift_lowpass_g = np.fft.ifftshift(fshift_lowpass_g)
img_back_lowpass_g = np.fft.ifft2(f_ishift_lowpass_g)
img_back_lowpass_g = np.abs(img_back_lowpass_g)

f_ishift_highpass_g = np.fft.ifftshift(fshift_highpass_g)
img_back_highpass_g = np.fft.ifft2(f_ishift_highpass_g)
img_back_highpass_g = np.abs(img_back_highpass_g)

f_ishift_bandpass_g = np.fft.ifftshift(fshift_bandpass_g)
img_back_bandpass_g = np.fft.ifft2(f_ishift_bandpass_g)
img_back_bandpass_g = np.abs(img_back_bandpass_g)

f_ishift_lowpass_r = np.fft.ifftshift(fshift_lowpass_r)
img_back_lowpass_r = np.fft.ifft2(f_ishift_lowpass_r)
img_back_lowpass_r = np.abs(img_back_lowpass_r)

f_ishift_highpass_r = np.fft.ifftshift(fshift_highpass_r)
img_back_highpass_r = np.fft.ifft2(f_ishift_highpass_r)
img_back_highpass_r = np.abs(img_back_highpass_r)

f_ishift_bandpass_r = np.fft.ifftshift(fshift_bandpass_r)
img_back_bandpass_r = np.fft.ifft2(f_ishift_bandpass_r)
img_back_bandpass_r = np.abs(img_back_bandpass_r)

# Normalisasi intensitas untuk setiap channel
img_back_lowpass_b = cv2.normalize(img_back_lowpass_b, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)
img_back_highpass_b = cv2.normalize(img_back_highpass_b, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)
img_back_bandpass_b = cv2.normalize(img_back_bandpass_b, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)

img_back_lowpass_g = cv2.normalize(img_back_lowpass_g, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)
img_back_highpass_g = cv2.normalize(img_back_highpass_g, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)
img_back_bandpass_g = cv2.normalize(img_back_bandpass_g, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)

img_back_lowpass_r = cv2.normalize(img_back_lowpass_r, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)
img_back_highpass_r = cv2.normalize(img_back_highpass_r, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)
img_back_bandpass_r = cv2.normalize(img_back_bandpass_r, None, 255, 0, cv2.NORM_MINMAX, cv2.CV_8U)

# Gabungkan kembali channel
img_back_lowpass = cv2.merge((img_back_lowpass_b, img_back_lowpass_g, img_back_lowpass_r))
img_back_highpass = cv2.merge((img_back_highpass_b, img_back_highpass_g, img_back_highpass_r))
img_back_bandpass = cv2.merge((img_back_bandpass_b, img_back_bandpass_g, img_back_bandpass_r))

# Tampilkan hasil dengan Matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(img_back_lowpass, cv2.COLOR_BGR2RGB))
plt.title('Low-Pass')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(img_back_highpass, cv2.COLOR_BGR2RGB))
plt.title('High-Pass')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(img_back_bandpass, cv2.COLOR_BGR2RGB))
plt.title('Band-Pass')
plt.axis('off')

plt.tight_layout()
plt.show()
```
